def calculate_mean(numbers):

# Function to find the median  
def calculate_median(numbers):  
    sorted_numbers = sorted(numbers)  
    n = len(sorted_numbers)  
    mid = n // 2  
    if n % 2 == 0:  
        return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2  
    else:  
        return sorted_numbers[mid]  

# (ii) Function to compute the mode  
def calculate_mode(numbers):  
    frequency = Counter(numbers)  
    mode_data = frequency.most_common()  
    max_freq = mode_data[0][1]  
    modes = [num for num, freq in mode_data if freq == max_freq]  
    return modes  

# (iii) Function to calculate the weighted mean  
def calculate_weighted_mean(values, weights):  
    return sum(value * weight for value, weight in zip(values, weights)) / sum(weights)  

# (iv) Function to find the geometric mean

    return gmean(numbers)  

# (v) Function to calculate the harmonic mean  
def calculate_harmonic_mean(numbers):  
    return hmean(numbers)  

# (vi) Function to determine the midrange  
def calculate_midrange(numbers):  
    return (min(numbers) + max(numbers)) / 2  

# (vii) Function to find the trimmed mean

    sorted_numbers = sorted(numbers)  
    trimmed_numbers = sorted_numbers[k:n-k]  # Trim the lowest and highest k elements

class DiscreteRandomVariable:  
    def __init__(self, values, probabilities):  
        if len(values) != len(probabilities):  
            raise ValueError("Length of values and probabilities must be the same.")  
        
        if any(p < 0 or p > 1 for p in probabilities):  
            raise ValueError("Probabilities must be between 0 and 1.")  
        
        if not abs(sum(probabilities) - 1) < 1e-10:  
            raise ValueError("Probabilities must sum to 1.")  

        self.values = values  
        self.probabilities = probabilities  

    def expected_value(self):  
        """Calculate and return the expected value of the random variable."""  
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))  

    def variance(self):  
        """Calculate and return the variance of the random variable."""  
        expected_value = self.expected_value()  
        return sum(prob * (value - expected_value) ** 2 for value, prob in zip(self.values, self.probabilities))

class FairDie:  
    def __init__(self):  
        self.values = [1, 2, 3, 4, 5, 6]  # Possible outcomes of a six-sided die  
        self.probabilities = [1/6] * 6  # All outcomes have equal probability  

    def expected_value(self):  
        """Calculate and return the expected value of the die roll."""  
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))  

    def variance(self):  
        """Calculate and return the variance of the die roll."""  
        expected_value = self.expected_value()  
        return sum(prob * (value - expected_value) ** 2 for value, prob in zip(self.values, self.probabilities))

import numpy as np  

def generate_samples_and_calculate_statistics(distribution, params, num_samples):  
    """  
    Generate random samples from a specified probability distribution and calculate their mean and variance.  

    Parameters:  
    - distribution: A string representing the distribution ('binomial' or 'poisson').  
    - params: A tuple containing parameters specific to the distribution:  
        - For 'binomial', provide (n, p) where n is the number of trials and p is the probability of success.  
        - For 'poisson', provide (lambda) where lambda is the average number of occurrences.  
    - num_samples: The number of random samples to generate.  

    Returns:  
    - samples: The generated random samples.  
    - mean: The mean of the samples.  
    - variance: The variance of the samples.  
    """  
    if distribution == 'binomial':  
        n, p = params  
        samples = np.random.binomial(n, p, num_samples)  
    elif distribution == 'poisson':  
        lam = params[0]  # Only one parameter for Poisson  
        samples = np.random.poisson(lam, num_samples)  
    else:  
        raise ValueError("Unsupported distribution. Use 'binomial' or 'poisson'.")  

     mean = np.mean(samples)  
     variance = np.var(samples)  

     return samples, mean, variance

    """  
    Calculate the skewness of a given column in the dataset.  
    
    :param data: DataFrame containing the dataset  
    :param column: Column name to calculate skewness for  
    :return: Skewness value  
    """  
    return data[column].skew()  

def assess_skewness(data, column):  
    """  
    Determine whether a column exhibits positive, negative, or approximately symmetric skewness.  
    
    :param data: DataFrame containing the dataset  
    :param column: Column name to assess skewness for  
    :return: String indicating the type of skewness  
    """  
    skewness = calculate_skewness(data, column)  
    if skewness > 0:  
        return "Positive skewness"  
    elif skewness < 0:  
        return "Negative skewness"  
    else:

    :param col1: First column name

    :return: Covariance value

    return data[col1].cov(data[col2])

import numpy as np  
import scipy.stats as stats  

def analyze_satisfaction_scores(branch_a_scores, branch_b_scores):  
    # Calculate means and standard deviations  
    mean_branch_a = np.mean(branch_a_scores)  
    mean_branch_b = np.mean(branch_b_scores)  
    std_branch_a = np.std(branch_a_scores, ddof=1)  # Sample standard deviation  
    std_branch_b = np.std(branch_b_scores, ddof=1)  # Sample standard deviation  
    n_branch_a = len(branch_a_scores)  
    n_branch_b = len(branch_b_scores)  

    # Calculate the pooled standard deviation  
    pooled_std = np.sqrt(((n_branch_a - 1) * std_branch_a**2 + (n_branch_b - 1) * std_branch_b**2) / (n_branch_a + n_branch_b - 2))  

    # Calculate the t-statistic  
    t_statistic = (mean_branch_a - mean_branch_b) / (pooled_std * np.sqrt(1/n_branch_a + 1/n_branch_b))  

    # Calculate degrees of freedom  
    df = n_branch_a + n_branch_b - 2  

    # Calculate the p-value for a two-tailed test  
    p_value = stats.t.sf(np.abs(t_statistic), df) * 2  # two-tailed  

    # Display results

````python
import numpy as np
from collections import Counter
from scipy.stats import gmean, hmean

# Generate a list of 100 integers between 90 and 130
np.random.seed(0)  # For reproducibility
int_list = np.random.randint(90, 131, size=100).tolist()

# (i) Function to calculate the mean
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

# Function to find the median
def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2
    else:
        return sorted_numbers[mid]

# (ii) Function to compute the mode
def calculate_mode(numbers):
    frequency = Counter(numbers)
    mode_data = frequency.most_common()
    max_freq = mode_data[0][1]
    modes = [num for num, freq in mode_data if freq == max_freq]
    return modes

# (iii) Function to calculate the weighted mean
def calculate_weighted_mean(values, weights):
    return sum(value * weight for value, weight in zip(values, weights)) / sum(weights)

# (iv) Function to find the geometric mean
def calculate_geometric_mean(numbers):
    return gmean(numbers)

# (v) Function to calculate the harmonic mean
def calculate_harmonic_mean(numbers):
    return hmean(numbers)

# (vi) Function to determine the midrange
def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

# (vii) Function to find the trimmed mean
def calculate_trimmed_mean(numbers, trim_percentage):
    n = len(numbers)
    k = int(n * trim_percentage)
    sorted_numbers = sorted(numbers)
    trimmed_numbers = sorted_numbers[k:n-k]  # Trim the lowest and highest k elements
    return calculate_mean(trimmed_numbers)

# Testing the functions
print("Integer List:", int_list)
print("Mean:", calculate_mean(int_list))
print("Median:", calculate_median(int_list))
print("Mode:", calculate_mode(int_list))

# For weighted mean: using arbitrary weights, for example
weights = np.random.rand(100)  # Generate 100 random weights
print("Weighted Mean:", calculate_weighted_mean(int_list, weights))

print("Geometric Mean:", calculate_geometric_mean(int_list))
print("Harmonic Mean:", calculate_harmonic_mean(int_list))
print("Midrange:", calculate_midrange(int_list))

# Trimmed mean example with 10% trimming
trim_percentage = 0.1
print("Trimmed Mean (10%):", calculate_trimmed_mean(int_list, trim_percentage))
````

````python
import numpy as np
import scipy.stats as stats

def analyze_satisfaction_scores(branch_a_scores, branch_b_scores):
    # Calculate means and standard deviations
    mean_branch_a = np.mean(branch_a_scores)
    mean_branch_b = np.mean(branch_b_scores)
    std_branch_a = np.std(branch_a_scores, ddof=1)  # Sample standard deviation
    std_branch_b = np.std(branch_b_scores, ddof=1)  # Sample standard deviation
    n_branch_a = len(branch_a_scores)
    n_branch_b = len(branch_b_scores)

    # Calculate the pooled standard deviation
    pooled_std = np.sqrt(((n_branch_a - 1) * std_branch_a**2 + (n_branch_b - 1) * std_branch_b**2) / (n_branch_a + n_branch_b - 2))

    # Calculate the t-statistic
    t_statistic = (mean_branch_a - mean_branch_b) / (pooled_std * np.sqrt(1/n_branch_a + 1/n_branch_b))

    # Calculate degrees of freedom
    df = n_branch_a + n_branch_b - 2

    # Calculate the p-value for a two-tailed test
    p_value = stats.t.sf(np.abs(t_statistic), df) * 2  # two-tailed

    # Display results
    print(f"Mean Satisfaction Score (Branch A): {mean_branch_a:.2f}")
    print(f"Mean Satisfaction Score (Branch B): {mean_branch_b:.2f}")
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-value: {p_value:.4f}")

# Data for satisfaction scores of both branches
branch_a_scores = [4, 5, 3, 4, 5, 4, 5, 3, 4, 4, 5, 4, 4, 3, 4, 5, 5, 4, 3, 4, 5, 4, 3, 5, 4, 4, 5, 3, 4, 5, 4]
branch_b_scores = [3, 4, 2, 3, 4, 3, 4, 2, 3, 3, 4, 3, 3, 2, 3, 4, 4, 3, 2, 3, 4, 3, 2, 4, 3, 3, 4, 2, 3, 4, 3]

# Run the analysis
analyze_satisfaction_scores(branch_a_scores, branch_b_scores)
````

````python
import numpy as np
import scipy.stats as stats

def analyze_intervention(pre_intervention_scores, post_intervention_scores):
    # Convert lists to numpy arrays for easier calculations
    pre_scores = np.array(pre_intervention_scores)
    post_scores = np.array(post_intervention_scores)

    # Calculate the differences
    differences = post_scores - pre_scores

    # Calculate the mean and standard deviation of the differences
    mean_difference = np.mean(differences)
    std_difference = np.std(differences, ddof=1)  # Sample standard deviation
    n = len(differences)

    # Calculate the t-statistic
    t_statistic = mean_difference / (std_difference / np.sqrt(n))

    # Calculate degrees of freedom
    df = n - 1

    # Calculate the p-value for a one-tailed test
    p_value = stats.t.sf(t_statistic, df)  # Since we're testing if the post scores are greater

    # Display results
    print(f"Mean Difference (Post - Pre): {mean_difference:.2f}")
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-value: {p_value:.4f}")

# Data for pre- and post-intervention scores
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Run the analysis
analyze_intervention(pre_intervention_scores, post_intervention_scores)
````

````python
import numpy as np
import scipy.stats as stats

def analyze_cholesterol_levels(existing_drug_levels, new_drug_levels):
    # Convert lists to numpy arrays for easier calculations
    existing_drug = np.array(existing_drug_levels)
    new_drug = np.array(new_drug_levels)

    # Calculate means and standard deviations
    mean_existing = np.mean(existing_drug)
    mean_new = np.mean(new_drug)
    std_existing = np.std(existing_drug, ddof=1)  # Sample standard deviation
    std_new = np.std(new_drug, ddof=1)  # Sample standard deviation
    n_existing = len(existing_drug)
    n_new = len(new_drug)

    # Calculate the pooled standard deviation
    pooled_std = np.sqrt(((n_existing - 1) * std_existing**2 + (n_new - 1) * std_new**2) / (n_existing + n_new - 2))

    # Calculate the t-statistic
    t_statistic = (mean_new - mean_existing) / (pooled_std * np.sqrt(1/n_new + 1/n_existing))

    # Calculate degrees of freedom
    df = n_existing + n_new - 2

    # Calculate the p-value for a one-tailed test
    p_value = stats.t.cdf(t_statistic, df)  # Since we're looking for a reduction in cholesterol levels

    # Display results
    print(f"Mean Cholesterol Level (Existing Drug): {mean_existing:.2f}")
    print(f"Mean Cholesterol Level (New Drug): {mean_new:.2f}")
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-value: {p_value:.4f}")

# Data for cholesterol levels for existing and new drugs
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

# Run the analysis
analyze_cholesterol_levels(existing_drug_levels, new_drug_levels)
````

````python
import numpy as np
import scipy.stats as stats

def ab_test(layout_a_clicks, layout_b_clicks):
    # Convert lists to numpy arrays for easier calculations
    layout_a = np.array(layout_a_clicks)
    layout_b = np.array(layout_b_clicks)

    # Calculate means and standard deviations
    mean_a = np.mean(layout_a)
    mean_b = np.mean(layout_b)
    std_a = np.std(layout_a, ddof=1)  # Sample standard deviation
    std_b = np.std(layout_b, ddof=1)  # Sample standard deviation
    n_a = len(layout_a)
    n_b = len(layout_b)

    # Calculate the t-statistic
    t_statistic = (mean_a - mean_b) / np.sqrt((std_a**2 / n_a) + (std_b**2 / n_b))

    # Calculate degrees of freedom
    df = ((std_a**2 / n_a + std_b**2 / n_b) ** 2) / (
        (std_a**2 / n_a)**2 / (n_a - 1) + (std_b**2 / n_b)**2 / (n_b - 1)
    )

    # Calculate the p-value for the two-tailed test
    p_value = stats.t.sf(np.abs(t_statistic), df) * 2  # two-tailed test

    # Display the results
    print(f"Mean Clicks (Layout A): {mean_a:.2f}")
    print(f"Mean Clicks (Layout B): {mean_b:.2f}")
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"Degrees of Freedom: {df:.2f}")
    print(f"P-value: {p_value:.4f}")

# Data for clicks in layouts A and B
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Run the A/B test analysis
ab_test(layout_a_clicks, layout_b_clicks)
````

````python
import numpy as np
import scipy.stats as stats

# Sample data
response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])

# Calculate the sample mean and standard deviation
sample_mean = np.mean(response_times)
sample_std = np.std(response_times, ddof=1)  # Sample standard deviation
n = len(response_times)

# Claimed population mean
claimed_mean = 5.0

# Calculate Z-score
z_score = (sample_mean - claimed_mean) / (sample_std / np.sqrt(n))

# Calculate the P-value for a one-tailed test
p_value = stats.norm.cdf(z_score)  # Since we're looking for a decrease

# Display results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std:.4f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

# Make a decision based on the p-value
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The average response time is significantly less than 5 minutes.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that the average response time is less than 5 minutes.")
````

````python
import numpy as np
import statsmodels.api as sm
import scipy.stats as stats

# Generate the data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Count of successes (purchases) and trials (visitors)
success_old = np.sum(old_layout)  # Number of purchases in old layout
n_old = len(old_layout)            # Total visitors in old layout

success_new = np.sum(new_layout)  # Number of purchases in new layout
n_new = len(new_layout)            # Total visitors in new layout

# Calculate conversion rates
p_old = success_old / n_old
p_new = success_new / n_new

print(f"Old Layout Conversion Rate: {p_old:.4f}")
print(f"New Layout Conversion Rate: {p_new:.4f}")

# Perform Z-test for proportions
z_score, p_value = sm.stats.proportions_ztest([success_old, success_new], [n_old, n_new], alternative='larger')

# Display the results
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

# Make a decision based on the p-value
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The new layout has a significantly higher conversion rate.")
else:
    print("Fail to reject the null hypothesis: No significant difference in conversion rates.")
````

````python
import numpy as np
import math

def poisson_pmf(k, rate):
    """
    Calculate the probability mass function (PMF) of a Poisson distribution.

    Parameters:
    - k: The number of occurrences (must be a non-negative integer).
    - rate: The average rate () at which events occur.

    Returns:
    - pmf_value: The probability mass function value for k occurrences.
    """
    if k < 0:
        raise ValueError("k must be a non-negative integer.")

    pmf_value = (rate**k * np.exp(-rate)) / math.factorial(k)
    return pmf_value

# Example usage
if __name__ == "__main__":
    # Define the rate parameter ()
    rate = 4.0  # Example rate (mean number of occurrences)

    # Evaluate PMF at specific values of k
    k_values = [0, 1, 2, 3, 4, 5, 6]
    for k in k_values:
        pmf_value = poisson_pmf(k, rate)
        print(f"PMF at k = {k}: {pmf_value:.4f}")
````

````python
import numpy as np

def exponential_cdf(x, rate):
    """
    Calculate the cumulative distribution function (CDF) of an exponential distribution.

    Parameters:
    - x: The value at which to evaluate the CDF (must be x >= 0).
    - rate: The rate parameter () of the exponential distribution.

    Returns:
    - cdf_value: The cumulative distribution function value at x.
    """
    if x < 0:
        return 0  # CDF is 0 for all x < 0
    cdf_value = 1 - np.exp(-rate * x)
    return cdf_value

# Example usage
if __name__ == "__main__":
    # Define the rate parameter ()
    rate = 1.5  # Example rate (can also be mean inverse)

    # Evaluate CDF at specific points
    x_values = [0, 1, 2, 3, 4, 5]
    for x in x_values:
        cdf_value = exponential_cdf(x, rate)
        print(f"CDF at x = {x}: {cdf_value:.4f}")
````

````python
import numpy as np

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    Parameters:
    - x: The value at which to evaluate the PDF.
    - mean: The mean () of the normal distribution.
    - std_dev: The standard deviation () of the normal distribution.

    Returns:
    - pdf_value: The probability density function value at x.
    """
    coeff = 1 / (std_dev * np.sqrt(2 * np.pi))
    exponent = -((x - mean)**2) / (2 * std_dev**2)
    pdf_value = coeff * np.exp(exponent)
    return pdf_value

# Example usage
if __name__ == "__main__":
    # Define parameters for the normal distribution
    mean = 0        # mean ()
    std_dev = 1     # standard deviation ()

    # Evaluate PDF at a specific point
    x_values = [-2, -1, 0, 1, 2]
    for x in x_values:
        pdf_value = normal_pdf(x, mean, std_dev)
        print(f"PDF at x = {x}: {pdf_value:.4f}")
````

````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

def calculate_skewness(data, column):
    """
    Calculate the skewness of a given column in the dataset.

    :param data: DataFrame containing the dataset
    :param column: Column name to calculate skewness for
    :return: Skewness value
    """
    return data[column].skew()

def assess_skewness(data, column):
    """
    Determine whether a column exhibits positive, negative, or approximately symmetric skewness.

    :param data: DataFrame containing the dataset
    :param column: Column name to assess skewness for
    :return: String indicating the type of skewness
    """
    skewness = calculate_skewness(data, column)
    if skewness > 0:
        return "Positive skewness"
    elif skewness < 0:
        return "Negative skewness"
    else:
        return "Approximately symmetric"

def calculate_covariance(data, col1, col2):
    """
    Calculate the covariance between two columns in the dataset.

    :param data: DataFrame containing the dataset
    :param col1: First column name
    :param col2: Second column name
    :return: Covariance value
    """
    return data[col1].cov(data[col2])

def calculate_pearson_correlation(data, col1, col2):
    """
    Calculate the Pearson correlation coefficient between two columns in the dataset.

    :param data: DataFrame containing the dataset
    :param col1: First column name
    :param col2: Second column name
    :return: Pearson correlation coefficient
    """
    return data[col1].corr(data[col2])

def plot_scatter(data, col1, col2):
    """
    Visualize the correlation between two specific columns using a scatter plot.

    :param data: DataFrame containing the dataset
    :param col1: First column name
    :param col2: Second column name
    """
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=data, x=col1, y=col2)
    plt.title(f'Scatter Plot of {col1} vs {col2}')
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.grid()
    plt.show()

if __name__ == "__main__":
    # Columns to analyze
    total_bill_col = 'total_bill'
    tip_col = 'tip'

    # Skewness computations
    skewness_total_bill = calculate_skewness(tips, total_bill_col)
    skewness_tip = calculate_skewness(tips, tip_col)
    print(f"Skewness of {total_bill_col}: {skewness_total_bill}")
    print(f"Skewness of {tip_col}: {skewness_tip}")

    print(f"{total_bill_col}: {assess_skewness(tips, total_bill_col)}")
    print(f"{tip_col}: {assess_skewness(tips, tip_col)}")

    # Covariance computation
    covariance = calculate_covariance(tips, total_bill_col, tip_col)
    print(f"Covariance between {total_bill_col} and {tip_col}: {covariance}")

    # Pearson correlation computation
    pearson_corr = calculate_pearson_correlation(tips, total_bill_col, tip_col)
    print(f"Pearson correlation coefficient between {total_bill_col} and {tip_col}: {pearson_corr}")

    # Scatter plot
    plot_scatter(tips, total_bill_col, tip_col)
````

````python
import numpy as np

def generate_normal_distribution_samples(mean, std_dev, num_samples):
    """
    Generate random samples from a normal (Gaussian) distribution.

    Parameters:
    - mean: The mean () of the normal distribution.
    - std_dev: The standard deviation () of the normal distribution.
    - num_samples: The number of random samples to generate.

    Returns:
    - samples: The generated random samples.
    """
    samples = np.random.normal(mean, std_dev, num_samples)
    return samples

def calculate_statistics(samples):
    """
    Calculate mean, variance, and standard deviation of the samples.

    Parameters:
    - samples: The random samples.

    Returns:
    - stats: A dictionary containing mean, variance, and standard deviation.
    """
    mean = np.mean(samples)
    variance = np.var(samples)
    std_dev = np.std(samples)

    stats = {
        'mean': mean,
        'variance': variance,
        'std_dev': std_dev
    }

    return stats

# Example usage
if __name__ == "__main__":
    mean = 0          # Mean of the normal distribution
    std_dev = 1      # Standard deviation of the normal distribution
    num_samples = 1000  # Number of samples to generate

    # Generate normal distribution samples
    samples = generate_normal_distribution_samples(mean, std_dev, num_samples)

    # Calculate statistics
    stats = calculate_statistics(samples)

    # Print results
    print(f"Generated {num_samples} samples from a normal distribution with mean {mean} and standard deviation {std_dev}.")
    print(f"Mean: {stats['mean']}")
    print(f"Variance: {stats['variance']}")
    print(f"Standard Deviation: {stats['std_dev']}")
````

````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

def calculate_skewness(data, column):
    """
    Calculate the skewness of a given column in the dataset.

    :param data: DataFrame containing the dataset
    :param column: Column name to calculate skewness for
    :return: Skewness value
    """
    return data[column].skew()

def assess_skewness(data, column):
    """
    Determine whether a column exhibits positive, negative, or approximately symmetric skewness.

    :param data: DataFrame containing the dataset
    :param column: Column name to assess skewness for
    :return: String indicating the type of skewness
    """
    skewness = calculate_skewness(data, column)
    if skewness > 0:
        return "Positive skewness"
    elif skewness < 0:
        return "Negative skewness"
    else:
        return "Approximately symmetric"

def calculate_covariance(data, col1, col2):
    """
    Calculate the covariance between two columns in the dataset.

    :param data: DataFrame containing the dataset
    :param col1: First column name
    :param col2: Second column name
    :return: Covariance value
    """
    return data[col1].cov(data[col2])

def calculate_pearson_correlation(data, col1, col2):
    """
    Calculate the Pearson correlation coefficient between two columns in the dataset.

    :param data: DataFrame containing the dataset
    :param col1: First column name
    :param col2: Second column name
    :return: Pearson correlation coefficient
    """
    return data[col1].corr(data[col2])

def plot_scatter(data, col1, col2):
    """
    Visualize the correlation between two specific columns using a scatter plot.

    :param data: DataFrame containing the dataset
    :param col1: First column name
    :param col2: Second column name
    """
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=data, x=col1, y=col2)
    plt.title(f'Scatter Plot of {col1} vs {col2}')
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.grid()
    plt.show()

if __name__ == "__main__":
    # Columns to analyze
    total_bill_col = 'total_bill'
    tip_col = 'tip'

    # Skewness computations
    skewness_total_bill = calculate_skewness(tips, total_bill_col)
    skewness_tip = calculate_skewness(tips, tip_col)
    print(f"Skewness of {total_bill_col}: {skewness_total_bill}")
    print(f"Skewness of {tip_col}: {skewness_tip}")

    print(f"{total_bill_col}: {assess_skewness(tips, total_bill_col)}")
    print(f"{tip_col}: {assess_skewness(tips, tip_col)}")

    # Covariance computation
    covariance = calculate_covariance(tips, total_bill_col, tip_col)
    print(f"Covariance between {total_bill_col} and {tip_col}: {covariance}")

    # Pearson correlation computation
    pearson_corr = calculate_pearson_correlation(tips, total_bill_col, tip_col)
    print(f"Pearson correlation coefficient between {total_bill_col} and {tip_col}: {pearson_corr}")

    # Scatter plot
    plot_scatter(tips, total_bill_col, tip_col)
````

````python
import numpy as np

def generate_samples_and_calculate_statistics(distribution, params, num_samples):
    """
    Generate random samples from a specified probability distribution and calculate their mean and variance.

    Parameters:
    - distribution: A string representing the distribution ('binomial' or 'poisson').
    - params: A tuple containing parameters specific to the distribution:
        - For 'binomial', provide (n, p) where n is the number of trials and p is the probability of success.
        - For 'poisson', provide (lambda) where lambda is the average number of occurrences.
    - num_samples: The number of random samples to generate.

    Returns:
    - samples: The generated random samples.
    - mean: The mean of the samples.
    - variance: The variance of the samples.
    """
    if distribution == 'binomial':
        n, p = params
        samples = np.random.binomial(n, p, num_samples)
    elif distribution == 'poisson':
        lam = params[0]  # Only one parameter for Poisson
        samples = np.random.poisson(lam, num_samples)
    else:
        raise ValueError("Unsupported distribution. Use 'binomial' or 'poisson'.")

    mean = np.mean(samples)
    variance = np.var(samples)

    return samples, mean, variance

# Example usage:
if __name__ == "__main__":
    # Generate samples from a binomial distribution
    binomial_samples, binomial_mean, binomial_variance = generate_samples_and_calculate_statistics(
        'binomial', (10, 0.5), 1000
    )
    print(f"Binomial distribution samples: {binomial_samples[:10]}")
    print(f"Mean: {binomial_mean}, Variance: {binomial_variance}")

    # Generate samples from a Poisson distribution
    poisson_samples, poisson_mean, poisson_variance = generate_samples_and_calculate_statistics(
        'poisson', (5,), 1000
    )
    print(f"\nPoisson distribution samples: {poisson_samples[:10]}")
    print(f"Mean: {poisson_mean}, Variance: {poisson_variance}")
````

````python
import numpy as np

class FairDie:
    def __init__(self):
        self.values = [1, 2, 3, 4, 5, 6]  # Possible outcomes of a six-sided die
        self.probabilities = [1/6] * 6  # All outcomes have equal probability

    def expected_value(self):
        """Calculate and return the expected value of the die roll."""
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
        """Calculate and return the variance of the die roll."""
        expected_value = self.expected_value()
        return sum(prob * (value - expected_value) ** 2 for value, prob in zip(self.values, self.probabilities))

# Simulating rolling the die multiple times
def simulate_rolls(num_rolls):
    die = FairDie()
    outcomes = [np.random.choice(die.values) for _ in range(num_rolls)]
    return outcomes

# Example usage
if __name__ == "__main__":
    num_rolls = 10000  # Number of simulated rolls
    outcomes = simulate_rolls(num_rolls)

    # Calculate expected value and variance
    fair_die = FairDie()
    print(f"Expected Value: {fair_die.expected_value()}")
    print(f"Variance: {fair_die.variance()}")

    # Optional: display the simulated outcomes for confirmation
    print(f"Sample of simulated outcomes: {outcomes[:10]}")  # Print first 10 rolls
````

````python
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        if len(values) != len(probabilities):
            raise ValueError("Length of values and probabilities must be the same.")

        if any(p < 0 or p > 1 for p in probabilities):
            raise ValueError("Probabilities must be between 0 and 1.")

        if not abs(sum(probabilities) - 1) < 1e-10:
            raise ValueError("Probabilities must sum to 1.")

        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        """Calculate and return the expected value of the random variable."""
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
        """Calculate and return the variance of the random variable."""
        expected_value = self.expected_value()
        return sum(prob * (value - expected_value) ** 2 for value, prob in zip(self.values, self.probabilities))

# Example usage:
values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]  # Example probabilities that sum to 1

random_variable = DiscreteRandomVariable(values, probabilities)
print(f"Expected Value: {random_variable.expected_value()}")
print(f"Variance: {random_variable.variance()}")
````

````python
def range_based_coefficient_of_dispersion(data):
    if not data:  # Check for an empty list
        return 0

    # Calculate the range
    data_range = max(data) - min(data)

    # Calculate the mean
    mean = sum(data) / len(data)

    # Calculate the coefficient of dispersion
    if mean == 0:  # Avoid division by zero
        return float('inf')  # Return infinity if mean is zero

    coefficient_of_dispersion = data_range / mean

    return coefficient_of_dispersion

# Example usage:
data = [10, 20, 30, 40, 50]
cd = range_based_coefficient_of_dispersion(data)
print(f"Range-Based Coefficient of Dispersion: {cd}")
````

````python
def calculate_coefficient_of_variation(data):
    if len(data) == 0:
        return None  # Return None if the dataset is empty

    mean = sum(data) / len(data)
    if mean == 0:
        return float('inf')  # Avoid division by zero

    # Calculate standard deviation
    variance = sum((x - mean) ** 2 for x in data) / len(data)
    standard_deviation = variance ** 0.5

    # Calculate coefficient of variation
    coefficient_of_variation = (standard_deviation / mean) * 100  # Convert to percentage
    return coefficient_of_variation

# Example usage:
data = [10, 12, 23, 23, 16, 23, 21, 16]
cv = calculate_coefficient_of_variation(data)
if cv is not None:
    print(f"Coefficient of Variation: {cv:.2f}%")
else:
    print("The dataset is empty.")
````

````python
import numpy as np

def compute_iqr(values):
    """
    Compute the interquartile range (IQR) of a list of values.

    Parameters:
        values (list or array-like): A list of numerical values.

    Returns:
        float: The interquartile range.
    """
    # Check if the input list is empty
    if not values:
        raise ValueError("The input list must not be empty.")

    # Convert values to a numpy array to handle statistical calculations
    values_array = np.array(values)

    # Calculate the first quartile (Q1) and third quartile (Q3)
    Q1 = np.percentile(values_array, 25)
    Q3 = np.percentile(values_array, 75)

    # Calculate the IQR
    IQR = Q3 - Q1

    return IQR

# Example usage
data = [6, 8, 12, 15, 18, 22, 24, 29, 30]
iqr = compute_iqr(data)
print("The Interquartile Range (IQR) is:", iqr)
````

````python
import numpy as np

def calculate_trimmed_mean(numbers, trim_percent):
    if not numbers:
        raise ValueError("The list cannot be empty.")

    if not (0 <= trim_percent < 50):
        raise ValueError("Trim percentage must be between 0 (inclusive) and 50 (exclusive).")

    # Calculate the number of elements to trim
    n = len(numbers)
    trim_count = int(n * trim_percent / 100)

    # Sort the numbers
    sorted_numbers = sorted(numbers)

    # Trim the sorted numbers
    trimmed_numbers = sorted_numbers[trim_count:n-trim_count]

    # Calculate the trimmed mean
    trimmed_mean = np.mean(trimmed_numbers) if trimmed_numbers else 0
    return trimmed_mean

# Example usage
number_list = [10, 20, 30, 40, 100, 200, 300]
trim_percent = 20  # Trim 20% of the lowest and 20% of the highest values
trimmed_mean = calculate_trimmed_mean(number_list, trim_percent)
print(f"Trimmed Mean: {trimmed_mean:.2f}")
````

````python
def calculate_harmonic_mean(values):
    if not values:
        raise ValueError("The list cannot be empty.")

    # Check for non-positive values
    if any(x <= 0 for x in values):
        raise ValueError("All values must be positive numbers greater than zero.")

    n = len(values)
    reciprocal_sum = sum(1 / x for x in values)  # Calculate the sum of the reciprocals

    # Calculate the harmonic mean
    harmonic_mean = n / reciprocal_sum
    return harmonic_mean

# Example usage
positive_values = [4, 5, 6]
harmonic_mean = calculate_harmonic_mean(positive_values)
print(f"Harmonic Mean: {harmonic_mean:.2f}")
````

````python
import math

def calculate_geometric_mean(numbers):
    if not numbers or any(num <= 0 for num in numbers):
        raise ValueError("The list must contain only positive numbers and cannot be empty.")

    product = 1
    for num in numbers:
        product *= num

    n = len(numbers)
    geometric_mean = math.pow(product, 1/n)  # Raise the product to the power of 1/n
    return geometric_mean

# Example usage
positive_numbers = [1, 3, 9, 27]
geometric_mean = calculate_geometric_mean(positive_numbers)
print(f"Geometric Mean: {geometric_mean:.2f}")
````

````python
def calculate_weighted_mean(values, weights):
    if len(values) != len(weights) or len(values) == 0:
        raise ValueError("Values and weights must have the same length and cannot be empty.")

    weighted_sum = sum(value * weight for value, weight in zip(values, weights))
    total_weight = sum(weights)

    if total_weight == 0:
        raise ValueError("The sum of weights cannot be zero.")

    return weighted_sum / total_weight

# Example usage
values = [10, 20, 30]
weights = [1, 2, 3]  # Weights for each corresponding value

weighted_mean = calculate_weighted_mean(values, weights)
print(f"Weighted Mean: {weighted_mean}")
````

````python
from collections import Counter

def calculate_mode(numbers):
    if not numbers:
        return None  # Return None for empty list

    # Count the frequency of each number in the list
    frequency = Counter(numbers)

    # Find the maximum frequency
    max_freq = max(frequency.values())

    # Find all numbers that have the maximum frequency
    modes = [num for num, freq in frequency.items() if freq == max_freq]

    # If all numbers are unique, there is no mode
    return modes if len(modes) < len(numbers) else None

# Example usage
number_list = [1, 2, 2, 3, 4, 4, 5, 5, 5]
mode = calculate_mode(number_list)
print(f"Mode: {mode}")
````

````python
def calculate_median(numbers):
    if not numbers:
        return 0  # Return 0 for empty list
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    mid = n // 2

    # If the count is even, the median is the average of the two middle numbers
    if n % 2 == 0:
        return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2
    else:
        return sorted_numbers[mid]

# Example usage:
median = calculate_median(numbers)
print(f"Median: {median}")
````
